```python
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import imdb
from keras.preprocessing import sequence
from IPython.display import display, Markdown

%matplotlib inline
```

```python
from keras import models
from keras import optimizers as opt
from keras.layers import Input
from keras.layers import Embedding
from keras.layers import LSTM
from keras.layers import Conv1D
from keras.layers import MaxPooling1D
from keras.layers import Dense
from attention.layers import AttentionLayer

# Please notice that this dataset is quite simple, so it's easy to overfit a model on it.
# We are using it because it comes bundled with Keras and the goal is to showcase the layer only
```

```python
VOCAB_SIZE = 10000
MAX_LEN = 100
HIDDEN_SIZE = 8
DROPOUT = 0.5

(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=VOCAB_SIZE)

x_train = sequence.pad_sequences(x_train, MAX_LEN, padding='post', truncating='post')
x_test = sequence.pad_sequences(x_test, MAX_LEN, padding='post', truncating='post')
```

```python
def build_model():
    sentence_in = Input((MAX_LEN, ),
                        name='sentence-in')

    embedded = Embedding(VOCAB_SIZE,
                         HIDDEN_SIZE,
                         mask_zero=False,
                         name='embedding')(sentence_in)

    conv = Conv1D(name='conv', filters=HIDDEN_SIZE, kernel_size=3, strides=1)(embedded)
    pool = MaxPooling1D(name='pool', pool_size=2)(conv)

    sentence = AttentionLayer(name='attention')(pool)

    output   = Dense(1,
                     activation='sigmoid',
                     name='output')(sentence)

    model = models.Model(inputs=[sentence_in], outputs=[output])
    return model
```

```python
# (b, t, d)
model = build_model()
model.summary()
    
model.compile('adam', 'binary_crossentropy', metrics=['binary_accuracy'])
model.fit(x_train, y_train, batch_size=32, epochs=2, validation_split=0.2)
```

```python
def build_viz_model(model):
    sentence_in = Input((MAX_LEN, ),
                        name='sentence-in')

    embedded = Embedding(VOCAB_SIZE,
                         HIDDEN_SIZE,
                         mask_zero=False,
                         weights=model.layers[1].get_weights(),
                         name='embedding')(sentence_in)

    conv = Conv1D(filters=HIDDEN_SIZE,
                  kernel_size=3,
                  strides=1,
                  weights=model.layers[2].get_weights(),
                  name='conv1d')(embedded)
    
    pool = MaxPooling1D(name='pool',
                        pool_size=2,
                        weights=model.layers[3].get_weights())(conv)

    alphas = AttentionLayer(return_attention=True,
                            weights=model.layers[4].get_weights(),
                            name='attention')(pool)
    
    linear_alphas = Dense(49, name='linear')(alphas)

    return models.Model(inputs=[sentence_in], outputs=[linear_alphas])
```

```python
viz_model = build_viz_model(model)
viz_model.summary()
```

```python
from keras import backend as K
import tensorflow as tf

samples = x_train[:2]

def compute_sentence_betas(viz_model, samples):
    inn = viz_model.layers[1].output
    out = viz_model.layers[-1].output

    grads = K.gradients(out, inn)

    # The Jacobian is dA/dw, ie: the derivative of the attention function wrt each dimension of the word vectors
    with tf.Session() as s:
        s.run(tf.global_variables_initializer())
        jacobian = s.run(grads, feed_dict={viz_model.input: samples})[0]

    # First we transform these vectors into numbers by considering the magnitude of the gradient vector for each word
    unnormalized_beta = np.linalg.norm(jacobian, axis=2)

    # Since the attention values must range between [0, 1] and their sum must be 1, we normalize these values
    beta_normalization = np.sum(unnormalized_beta, axis=1, keepdims=True)

    # Then we normalize beta and call them the word attention weights
    beta = unnormalized_beta / beta_normalization
    return beta
```

```python
def get_index2word():
    """Computes the table that maps ids to their words."""
    INDEX_FROM = 3   # word index offset

    word2index = imdb.get_word_index()
    index2word = {v + INDEX_FROM: k for k, v in word2index.items()}
    index2word[0] = '[PAD]'
    index2word[1] = '[START]'
    index2word[2] = '[UNK]'

    return index2word


def reconstruct(sample, index2word):
    """Given a list of word ids, returns a list of words."""
    return [index2word[word] for word in sample]


def _weight2color(brightness):
    """Converts a single (positive) attention weight to a shade of blue."""
    brightness = brightness.item()

    brightness = int(round(255 * brightness)) # convert from 0.0-1.0 to 0-255
    ints = (255 - brightness, 255 - brightness, 255)
    return 'rgba({}, {}, {}, 0.6)'.format(*ints)


def print_sentence(label, predicted, sentence, weights):
    """Prints a sample (sequence) making the most attended words background darker."""

    parts = list()
    parts.append('<span style="padding:2px;">[actual: %10s >< pred: %10s]</span> ' % (label, predicted))
    for word, weight in zip(sentence, weights):
        if word == '[PAD]':
            break
        parts.append('<span style="background: {}; color:#000; padding:2px; font-weight=\'bold\'">{}</span>'.format(_weight2color(weight), word))
    
    text = ' '.join(parts)
    display(Markdown(text))


def plot_sentence(words, weights):
    words = [f'{i}_{word}' for i, word in enumerate(words)]
    
    plt.figure(figsize=(20, 2))
    plt.plot(words, z)
    plt.xticks(rotation=90)
    plt.grid(alpha=0.4)
    plt.ylabel('Attention')
    plt.show()
```

```python
index2word = get_index2word()
index2label = {
    0: 'negative',
    1: 'positive'
}

viz_model = build_viz_model(model)

for index in range(5):
    # reconstructing sample from word ids to actual words
    words = reconstruct(x_test[index], index2word)
    sample = x_test[index:index+1]

    # getting prediction and alphas
    pred = int(model.predict(sample)[0] >= 0.5)
    
    z = compute_sentence_betas(viz_model, x_test[index:index+1])[0]
    
    # reescaling for visualization purposes
    w = (z - np.min(z)) / (np.max(z) - np.min(z))

    # ta-da
    print_sentence(index2label[y_test[index]], index2label[pred], words, w)
    plot_sentence(words, z)
```
